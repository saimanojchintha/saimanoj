```python
left = pd.DataFrame({'Key': ['K0', 'K1', 'K2', 'K3'], 
                    'A': ['A0', 'A1', 'A2', 'A3'], 
                    'B': ['B0', 'B1', 'B2', 'B3']}) 

right = pd.DataFrame({'Key': ['K0', 'K1', 'K4', 'K5'], 
                    'C': ['C0', 'C1', 'C2', 'C3'], 
                    'D': ['D0', 'D1', 'D2', 'D3']}) 
left
right
```

```python
df3=pd.merge(left, right, on='Key', how='inner')
df3
```

```python
df3=pd.merge(left, right, on='Key', how='outer')
df3
```

```python
df3=pd.merge(left, right, on='Key', how='left')
df3
```

```python
df3=pd.merge(left, right, on='Key', how='right')
df3
```
