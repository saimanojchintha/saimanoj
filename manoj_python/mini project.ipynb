```python
import pandas as pd
import mysql.connector
from mysql.connector import errorcode

# Replace these values with your actual MySQL database credentials
host = 'localhost'
user = 'root'
password = 'sairam'
database = 'saimanoj'

# Create a connection to the MySQL database
try:
    connection = mysql.connector.connect(host=host, user=user, password=password, database=database)
    cursor = connection.cursor()
    print("Connected to MySQL database")

    # Define the table name in the MySQL database
    table_name = 'student_grade'

    # Create the table if it doesn't exist (optional)
    create_table_query = f'''
    CREATE TABLE IF NOT EXISTS {table_name} (
        id INT AUTO_INCREMENT PRIMARY KEY,
        name VARCHAR(255),
        score int(10)
    );
    '''
    cursor.execute(create_table_query)

    while True:
        print("\nOptions:")
        print("1. Select table")
        print("2. Insert values")
        print("3. Update table")
        print("4. Exit")

        choice = input("Enter your choice (1-4): ")

        if choice == '1':
            # Select table
            select_query = f"SELECT * FROM {table_name}"
            cursor.execute(select_query)
            rows = cursor.fetchall()

            if not rows:
                print("Table is empty.")
            else:
                print("\nTable Content:")
                for row in rows:
                    print(row)

        elif choice == '2':
            # Insert values
            ide = int(input("Enter your id: "))
            name = input("Enter name: ")
            score = int(input("Enter your score: "))
            
            insert_query = f"INSERT INTO {table_name} (id, name, score) VALUES (%s, %s, %s)"
            values = (ide, name, score)
            
            cursor.execute(insert_query, values)
            connection.commit()
            print(f"Inserted a row into the table '{table_name}'")

        elif choice == '3':
            # Update table
            ide = int(input("Enter the id of the record to update: "))
            new_name=int(input('if you want to update your name then modify: '))
            new_score = int(input("Enter the new score: "))
            
            update_query = f"UPDATE {table_name} SET score = %s WHERE id = %s"
            update_values = (new_score, ide)
            
            cursor.execute(update_query, update_values)
                connection.commit()
            print(f"Updated the score for record with id {ide}")

        elif choice == '4':
            print("Exiting program.")
            break
        else:
            print("Invalid choice. Please enter a number between 1 and 4.")

except mysql.connector.Error as err:
    if err.errno == errorcode.ER_ACCESS_DENIED_ERROR:
        print("Access denied. Please check your credentials.")
    elif err.errno == errorcode.ER_BAD_DB_ERROR:
        print("Database does not exist. Please create the database first.")
    else:
        print(f"Error: {err}")
finally:
    # Close the cursor and connection
    if 'cursor' in locals() and cursor:
        cursor.close()
    if 'connection' in locals() and connection.is_connected():
        connection.close()
        print("MySQL connection closed")
```
