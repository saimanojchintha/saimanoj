```python
import pandas as pd
import numpy as nm
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv(r'C:\Users\sairam\Downloads\diamonds.csv')
```

```python
data.head()
```

```python
data=data.drop("Unnamed: 0",axis=1)
data.head()
```

```python
plt.figure(figsize=(5,5))
plt.hist(data["price"],bins=150)
plt.title("frequency prices: price of diamond in carats\n", fontdict={"fontsize":20,"fontweight":100,"color":"black"})

plt.xlabel("price", fontdict={"fontsize":15,"fontweight":200,"color":"purple"})
plt.ylabel("count",fontdict={"fontsize":10,"fontweight":200,"color":"black"})
plt.show()
```

```python
group=data.groupby("cut").count()
group
```

```python
x_vals=[]
y_vals=[]
for i in [0,1,4,3,2]:
    x_vals.append(group.index[i])
    y_vals.append(group.iloc[i,0])
```

```python
print(x_vals)
print(y_vals)
```

```python
plt.bar(x_vals,y_vals)
plt.title("Frequency Plot: Cut of Diamonds\n", fontdict={'fontsize': 10, 'fontweight' : 20, 'color' : 'Gray'})
plt.xlabel("Cut", fontdict={'fontsize': 12, 'fontweight' : 20, 'color' : 'purple'})
plt.ylabel("Count", fontdict={'fontsize': 12, 'fontweight' : 20, 'color' : 'violet'})

plt.show()
```

```python
explode=(0,0,0.1,0,0)
plt.figure(figsize=[5,15])
plt.pie(y_vals,explode=explode,labels=x_vals,autopct='%1.0f%%',textprops={'fontsize': 10, 'fontweight' : 200, 'color' : 'Black'}, startangle=90)

plt.title("Distribution based on : Quality of cut\n", fontdict={'fontsize': 20, 'fontweight' : 20, 'color' : 'blue'})

plt.show()
```

```python
data_fair = data[data['cut'] == 'Fair']
data_good = data[data['cut'] == 'Good']
data_vgood = data[data['cut'] == 'Very Good']
data_premium = data[data['cut'] == 'Premium']
data_ideal = data[data['cut'] == 'Ideal']
```

```python
plt.boxplot([data_fair['price'], data_good['price'], data_vgood['price'], data_premium['price'], data_ideal['price']])
plt.show()
```

```python
data_fair = data[(data['cut'] == 'Fair') & (data['price'] < 12500) ]
data_good = data[(data['cut'] == 'Good') & (data['price'] < 12500)]
data_vgood = data[(data['cut'] == 'Very Good') & (data['price'] < 12500)]
data_premium = data[(data['cut'] == 'Premium') & (data['price'] < 12500)]
data_ideal = data[(data['cut'] == 'Ideal') & (data['price'] < 12500)]
```

```python
plt.figure(figsize = [16,10])
plt.boxplot([data_fair['price'], data_good['price'], data_vgood['price'], data_premium['price'], data_ideal['price']])

plt.xticks((1,2,3,4,5), ['Fair', 'Good', 'Very Good', 'Premium', 'Ideal'])

# Adding and formatting title
plt.title("Price vs Type of Cut\n", fontdict={'fontsize': 20, 'fontweight' : 20, 'color' : 'Green'})

# Labeling Axes
plt.xlabel("Type of Cut", fontdict={'fontsize': 12, 'fontweight' : 20, 'color' : 'Brown'})
plt.ylabel("Price", fontdict={'fontsize': 12, 'fontweight' : 20, 'color' : 'Brown'})

plt.show()
```
