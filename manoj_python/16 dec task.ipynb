```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv(r'C:\Users\sairam\Downloads\Company_sales.csv')
```

```python
#1.	Read toothpaste sales data of each month and show it using a scatter plotAlso, add a grid in the plot. gridline style should .


plt.scatter(data=df, x="toothpaste", y="month_number", alpha=1, marker='x')
plt.title("sales of toothpaste per month\n", fontdict={'fontsize': 20, 'fontweight' : 20, 'color' : 'black'})
plt.xlabel('Type of toothpaste ', fontdict={'fontsize': 10, 'fontweight' : 10, 'color' : 'Brown'})
plt.ylabel('months', fontdict={'fontsize': 10, 'fontweight' : 10, 'color' : 'Brown'})

plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)

plt.show()

```

```python
#2 Read sales data of bathing soap of all months and show it using a bar chart.
b=df['bathingsoap'].tolist()
m=df['month_number'].tolist()
plt.bar(m,b,label='bathing soaps in term
plt.title('sales of bathing soaps in all months\n', fontdict={'fontsize':10, 'fontweight':10,'color':'gray'})
plt.xlabel('types of bathingSoaps', fontdict={'fontsize':10, 'fontweight':10,'color':'gray'})
plt.ylabel('months', fontdict={'fontsize':10, 'fontweight':10,'color':'gray'})
plt.grid(True,linewidth=1,linestyle="--")
plt.legend(loc='upper left')
plt.show()
```

```python
#3.	Calculate total sale data for last year for each product and show it using a Pie chart
x_label = ['FaceCream', 'FaseWash', 'ToothPaste', 'Bathing soap', 'Shampoo', 'Moisturizer']
y_label   = [df ['facecream'].sum(), df ['facewash'].sum(), df ['toothpaste'].sum(), 
         df ['bathingsoap'].sum(), df ['shampoo'].sum(), df ['moisturizer'].sum()]
plt.pie(y_label, labels=x_label, autopct='%1.1f%%',textprops={},startangle=90)
plt.title('Sales data')
plt.show()
```
