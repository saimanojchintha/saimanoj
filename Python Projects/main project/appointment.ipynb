```python
from tabulate import tabulate 
import pandas as pd
import mysql.connector

connection = mysql.connector.connect(
    host='localhost',
    user='root',
    password='sairam',
    database='mainproject_db',
)

if connection.is_connected():
    print('Connection established')
    cursor = connection.cursor()

    def add_appointment():
        try:
            appoint_date = str(input("Enter your appointment date: "))
            opd_no = int(input("Enter your opdno: "))
            p_name = input("Enter patient name: ")
            doc_name = input("Enter doctor name: ")
            symp = input("Enter your symptom: ")
            treat = input("Treatment for your symptom is: ")

            insert_query = "INSERT INTO appoint (doa, opdno, pname, dname, symptom, treatement) VALUES (%s, %s, %s, %s, %s, %s)"
            values = (appoint_date, opd_no, p_name, doc_name, symp, treat)

            cursor.execute(insert_query, values)
            connection.commit()
            print(f"Inserted a row into the table appoint")
        except mysql.connector.Error as err1:
            print(f"MySQL Error: {err1}")
        except Exception as e1:
            print(f"Exception: {e1}")

    def update_appointment():
        try:
            opd_no = int(input("Enter the opdno of the appointment to update: "))
            p_name = input("Enter the patient name of the appointment to update: ")

            select_query = "SELECT opdno FROM appoint WHERE opdno = %s AND pname = %s"
            cursor.execute(select_query, (opd_no, p_name))
            existing_record = cursor.fetchone()

            if existing_record:
                appoint_date = str(input("Enter the updated appointment date (YYYY-MM-DD): "))
                doc_name = str(input("Enter the updated doctor name: "))

                update_query = "UPDATE appoint SET doa = %s, dname = %s WHERE opdno = %s"
                update_values = (appoint_date, doc_name, opd_no)

                cursor.execute(update_query, update_values)
                connection.commit()
                print(f"Updated the appointment with opdno {opd_no}")
            else:
                print('No such appointment exists for the provided opdno and patient name.')
        except mysql.connector.Error as err:
            print(f"MySQL Error: {err}")
        except Exception as e:
            print(f"Exception: {e}")
            
    def view_existing_appointment_records():
        try:
        # Fetch and display existing records from the 'doctor' table
            select_all_query = "SELECT * FROM appoint"
            cursor.execute(select_all_query)
            existing_records = cursor.fetchall()

            if existing_records:
                headers = ["Date of appointmnt","opd no","Patient name", "Doctor Name", "Symptom","Treatement"]
                data = [list(record) for record in existing_records]

                table = tabulate(data, headers=headers, tablefmt="grid")
                print(table)
            else:
                print("No existing records in the 'appoint' table.")
        except mysql.connector.Error as err:
            print(f"Error: {err}")
        except Exception as e:
            print(f"Exception: {e}")
    
```
