```python
from tabulate import tabulate 
import pandas as pd
import mysql.connector
connection=mysql.connector.connect(
host='localhost',
user='root',
password='sairam',
database='mainproject_db',
)
if connection.is_connected():
    print('connection established')
    cursor=connection.cursor()
    def add_doc():
        try:
            idno = int(input("Enter your didno: "))
            doc_name = input("Enter doctor name: ")
            doc_department = input("Enter department name: ")

            insert_query = "INSERT INTO doctor (did, dname, deparment) VALUES (%s, %s, %s)"
            values = (idno, doc_name, doc_department)

            cursor.execute(insert_query, values)
            connection.commit()
            print(f"Inserted a row into the table doctor")
        except mysql.connector.Error as err1:
            print(f"MySQL Error: {err1}")
        except Exception as e1:
            print(f"Exception: {e1}")

    def edit_doc():
        try:
            idno = int(input("Enter the didno of the record to update: "))
            select_query = "SELECT did FROM doctor WHERE did = %s"
            cursor.execute(select_query, (idno,))
            existing_record = cursor.fetchone()

            if existing_record:
                #new_doc_name = input("Enter the new doctor name: ")
                new_doc_department = input("Enter the new doctor department: ")
        
                update_query = "UPDATE doctor SET deparment = %s WHERE did = %s"
                update_values = (new_doc_department, idno)
        
                cursor.execute(update_query, update_values)
                connection.commit()
                print(f"Updated the record for did {idno}")
            else:
                print('No such opdno exists')
        except mysql.connector.Error as err:
            print(f"Error: {err}")
        except Exception as e:
            print(f"Exception: {e}")
            
    def delete_doc():
        try:
            didno_to_delete = int(input("Enter the doc_no to delete: "))
            select_query = "SELECT did FROM doctor WHERE did = %s"
            cursor.execute(select_query, (didno_to_delete,))
            existing_record = cursor.fetchone()

            if existing_record:
                delete_query = "DELETE FROM doctor WHERE did = %s"
                cursor.execute(delete_query, (didno_to_delete,))
                connection.commit()
                print(f"Deleted the record for did {didno_to_delete}")
            else:
                print('No such did exists')
        except mysql.connector.Error as err2:
            print(f"Error: {err2}")
        except Exception as e2:
            print(f"Exception: {e2}")
            
    def search_doc():
        try:
            didno_to_search = int(input("Enter the doc_id to search: "))
    
        # Check if the record with the given opdno exists
            select_query = "SELECT * FROM doctor WHERE did = %s"
            cursor.execute(select_query, (didno_to_search,))
            existing_record = cursor.fetchone()

            if existing_record:
                headers = ["idno","doc_name","doc_department"]
                data = [existing_record]

                table = tabulate(data, headers=["did", "dname", "deparment"], tablefmt="grid")
                print(table)
            else:
                print('No such opdno exists')
        except mysql.connector.Error as err:
            print(f"Error: {err}")
        except Exception as e:
            print(f"Exception: {e}")
            
    def view_existing_doctor_records():
        try:
        # Fetch and display existing records from the 'doctor' table
            select_all_query = "SELECT * FROM doctor"
            cursor.execute(select_all_query)
            existing_records = cursor.fetchall()

            if existing_records:
                headers = ["Doctor Id", "Doctor Name", "Department"]
                data = [list(record) for record in existing_records]

                table = tabulate(data, headers=headers, tablefmt="grid")
                print(table)
            else:
                print("No existing records in the 'doctor' table.")
        except mysql.connector.Error as err:
            print(f"Error: {err}")
        except Exception as e:
            print(f"Exception: {e}")
    
```
