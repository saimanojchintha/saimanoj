```python
from tabulate import tabulate 
import mysql.connector

connection = mysql.connector.connect(
    host='localhost',
    user='root',
    password='sairam',
    database='mainproject_db',
)

if connection.is_connected():
    print('connection established')
    cursor = connection.cursor()
    def get_unique_opdno():
        while True:
            opdno = int(input("Enter your opdno: "))
            cursor.execute("SELECT * FROM opd WHERE opdno = %s", (opdno,))
            existing_record = cursor.fetchone()
            if existing_record:
                print(f"Opdno {opdno} already exists. Please choose a different opdno.")
            else:
                return opdno

    def new_opd():
        try:
            # Get a unique opdno
            opdno = get_unique_opdno()

            # Fetch patient name from opd table
            cursor.execute("SELECT pname FROM opd WHERE opdno = %s", (opdno,))
            existing_record = cursor.fetchone()
            pat_name = existing_record[0] if existing_record else input("Enter patient name: ")

            # Get user input for appointment details
            doa = input("Enter appointment date (YYYY-MM-DD): ")

            # Fetch patient address and mobile number from patient table
            cursor.execute("SELECT padd, pmob FROM patient WHERE pname = %s", (pat_name,))
            existing_record = cursor.fetchone()
            padd = existing_record[0] if existing_record else input("Enter patient address: ")
            pmob = existing_record[1] if existing_record else input("Enter patient mobile number: ")

            # Fetch and display doctor list from doctor table
            cursor.execute("SELECT dname FROM doctor")
            doctors_list = [row[0] for row in cursor.fetchall()]

            # Displaying doctor list
            print("Doctor List:")
            for i, doctor in enumerate(doctors_list, 1):
                print(f"{i}. {doctor}")

            # Selecting doctor name from the displayed list
            doctor_choice = int(input("Select the doctor (enter the number): "))
            selected_doctor = doctors_list[doctor_choice - 1]

            # Get user input for symptom
            symptom = input("Enter symptom: ")

            # Fetch fee from opd table
            cursor.execute("SELECT fee FROM opd WHERE opdno = %s", (opdno,))
            existing_fee_record = cursor.fetchone()
            doc_fee = existing_fee_record[0] if existing_fee_record else input("Enter doctor fee: ")

            # Insert data into opd table
            opd_insert_query = "INSERT INTO opd (opdno, pname, dname, fee) VALUES (%s, %s, %s, %s)"
            opd_values = (opdno, pat_name, selected_doctor, doc_fee)
            cursor.execute(opd_insert_query, opd_values)
            connection.commit()
            print(f"Inserted a row into the table opd")

            # Insert data into appoint table
            appoint_insert_query = "INSERT INTO appoint (opdno, pname, dname, doa, symptom) VALUES (%s, %s, %s, %s, %s)"
            appoint_values = (opdno, pat_name, selected_doctor, doa, symptom)
            cursor.execute(appoint_insert_query, appoint_values)
            connection.commit()
            print(f"Inserted a row into the table appoint")

            # Insert data into patient table
            patient_insert_query = "INSERT INTO patient (pname, padd, pmob) VALUES (%s, %s, %s)"
            patient_values = (pat_name, padd, pmob)
            cursor.execute(patient_insert_query, patient_values)
            connection.commit()
            print("Inserted a row into the table patient")

        except mysql.connector.Error as err1:
            print(f"MySQL Error: {err1}")

        except Exception as e1:
            print(f"Exception: {e1}")
    def edit_opd():
        try:
       

            idno = int(input("Enter the opdno of the record to update: "))
            select_query = "SELECT opdno FROM opd WHERE opdno = %s"
            cursor.execute(select_query, (idno,))
            existing_record = cursor.fetchone()

            if existing_record:
                new_pat_name = input("Enter the new patient name: ")
                new_doc_name = input("Enter the new doctor name: ")
                new_doc_fee = int(input("Enter the new doctor fee: "))
    
                update_query = "UPDATE opd SET pname = %s, dname = %s, fee = %s WHERE opdno = %s"
                update_values = (new_pat_name, new_doc_name, new_doc_fee, idno)
    
                cursor.execute(update_query, update_values)
                connection.commit()
                print(f"Updated the record for opdno {idno}")
            else:
                print('No such opdno exists')
        except mysql.connector.Error as err:
            print(f"Error: {err}")
        except Exception as e:
            print(f"Exception: {e}")
    def delete_opd():
        try:
        
            opdno_to_delete = int(input("Enter the opdno of the record to delete: "))
            select_query = "SELECT opdno FROM opd WHERE opdno = %s"
            cursor.execute(select_query, (opdno_to_delete,))
            existing_record = cursor.fetchone()

            if existing_record:
                delete_query = "DELETE FROM opd WHERE opdno = %s"
                cursor.execute(delete_query, (opdno_to_delete,))
                connection.commit()
                print(f"Deleted the record for opdno {opdno_to_delete}")
            else:
                print('No such opdno exists')
        except mysql.connector.Error as err2:
            print(f"Error: {err2}")
        except Exception as e2:
            print(f"Exception: {e2}")
    def search_opd():
        try:
        
            opdno_to_search = int(input("Enter the opdno of the record to search: "))
    
        # Check if the record with the given opdno exists
            select_query = "SELECT * FROM opd WHERE opdno = %s"
            cursor.execute(select_query, (opdno_to_search,))
            existing_record = cursor.fetchone()

            if existing_record:
                headers = ["OpdNo", "Patient Name", "Doctor Name", "Doctor Fee"]
                data = [existing_record]

                table = tabulate(data, headers=["opdno", "pname", "dname", "fee"], tablefmt="grid")
                print(table)
            else:
                print('No such opdno exists')
        except mysql.connector.Error as err:
            print(f"Error: {err}")
        except Exception as e:
            print(f"Exception: {e}")
    def view_existing_opd_records():
        try:
        # Fetch and display existing records from the 'opd' table
            select_all_query = "SELECT * FROM opd"
            cursor.execute(select_all_query)
            existing_records = cursor.fetchall()

            if existing_records:
                headers = ["OpdNo", "Patient Name", "Doctor Name", "Doctor Fee"]
                data = [list(record) for record in existing_records]

                table = tabulate(data, headers=headers, tablefmt="grid")
                print(table)
            else:
                print("No existing records in the 'opd' table.")
        except mysql.connector.Error as err:
            print(f"Error: {err}")
        except Exception as e:
            print(f"Exception: {e}")
```
